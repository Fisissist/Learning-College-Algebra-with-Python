# Solving for X

```python
import sympy
from sympy import symbols
from sympy.solvers import solve
from sympy.parsing.mathematica import parse_mathematica

x = symbols('x')

try:
  eq = input('Enter equation: 0 = ')
  print("x = ", solve(eq,x))
except:
  print("Error, enter again.")
  eq = input('Enter equation: 0 = ')

  eq = parse_mathematica(eq)

  print("x = ", solve(eq,x))
```

Learing about the symbolic library, importing that library and specific funtions from that library. Using the solver to solve functions when equal to zero.

Learned about the parse funtion to sperate inputs so they are not treated as a single expresssion, such as implicit multiplication.
