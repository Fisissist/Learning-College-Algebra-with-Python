####1
Classify each of the following costs as fixed costs, variable costs, or blended costs. If a cost is blended, separate it into its
fixed and variable components.
  * a. Natural gas bill for 15 per month plus 0.33/m^2 .

  * b. A chief executive officer salary of 240,000 per year.

  * c. An author earning a royalty of 5% of sales.

  * d. Placing a commercial on television for 300,000.

  * e. A cellphone bill for 40 per month plus 0.25/minute for long distance.

  * f. Hourly production worker wages of 18/hr.

  * g. Sales staff who are compensated at a salary of 1,000 per month plus 15% of sales




```python
from sympy import symbols, core

Fixed: dict = {}
Variable: dict = {}
Blended: dict = {}

x = symbols('x', real=True)

costs = {"a":15 + .33*x, "b":240000, "c":x*.05, "d":300000, "e":40 + .25*x, "f":18*x, "g":1000 + .15*x}

print(costs)

# sorts cost into other dict
for cost, value in costs.items():
  if isinstance(value, int):
    Fixed[cost] = value
  elif isinstance(value, core.mul.Mul):
    Variable[cost] = value
  else:
    Blended[cost] = value

print(Fixed.keys())
print(Variable.keys())
print(Blended.keys())
```

```python
import matplotlib.pyplot as plt
import numpy as np

fixMonthly: float = 1500
costPerPage: float = 1.5
additionalPageCost: float = .02
studentCost:float = .1
units: float = 100

totVarRev = units * (costPerPage + additionalPageCost + studentCost)

xmin = 0
xmax = 300
ymin = 0
ymax = 50000

points = xmax

x = np.linspace(xmin,xmax,points)
y = totVarRev * x + fixMonthly

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax])

ax.set_ylabel('Income')
ax.set_xlabel('For every 100 Copies')
ax.set_title('Income per Copy Made')
ax.grid(True)

plt.plot(x,y,'b',label='Rocket')

print(f"The slope, or net income rate of change, is the Total Variable Revenue: ${totVarRev},\nand the y-intercept is the Fixed Monthly Lease: ${fixMonthly}.")

plt.show()

print(f'The monthly net income if, on average, the shop makes 25,000 copies for students each month is ${totVarRev * 250 + fixMonthly}')
```

####11image.png

```python
import matplotlib.pyplot as plt

totFixCost: float = 19000
uniVarCost: float = 37.5

products: list = ['A', 'B', 'C', 'D']
forecastedUnits: list = [1400, 1850, 2500, 2750]
sellPrices: list = [50, 48, 46, 44]

netIncomes: list = []
totVarCosts: list = []
totRevs: list = []

for i in range(len(forecastedUnits)):
    totVarCost = forecastedUnits[i] * uniVarCost
    totRev = forecastedUnits[i] * sellPrices[i]
    netInc = totRev - totFixCost

    totVarCosts.append(totVarCost)
    totRevs.append(totRev)
    netIncomes.append(netInc)

    print(products[i], end='||')
    print(f'Total Variable Cost: {totVarCost}', end='||')
    print(f'Total Revenue: {totRev}', end='||')
    print(f'Net Income: {netInc}')

plt.title('Total Variable Costs of Products')
plt.xlabel('Products')
plt.ylabel('Total Variable Costs')
plt.bar(products, totVarCosts)
plt.show()

plt.title('Total Revenues of Products')
plt.xlabel('Products')
plt.ylabel('Total Revenues')
plt.bar(products, totRevs)
plt.show()

plt.title('Net Income of Products')
plt.xlabel('Products')
plt.ylabel('Net Income')
plt.bar(products, netIncomes)
plt.show()
```
