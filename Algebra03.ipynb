# Defining a python function

```python
def f(x):
  y = 4 * x + 3
  return y

print(5,"\t",f(5))

for x in range(10):
  print(x,"\t",f(x))
```

# Plotting

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = range(10)
y = [f(i) for i in x]

ax.plot(x,y)
```

## Plotting and tabling

```python
import matplotlib.pyplot as plt

xmin = -4
xmax = 10
ymin = -10
ymax = 50

fig, ax = plt.subplots()

plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax],[0,0],'b')
plt.plot([0,0],[ymin,ymax],'b')

for x in range(xmin+1,xmax):
  y = 4 * x + 3
  plt.plot([x],[y],'r+')
  plt.text(x,y,str(x))
  print(x,"\t",y)

plt.show()
```

# Arrays

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax],[0,0],'k')
plt.plot([0,0],[ymin,ymax],'k')

y = 2*x**3 +1
plt.plot(x,y,'r')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

# setting plot min and max dimension
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax],[0,0],'k')
plt.plot([0,0],[ymin,ymax],'k')

# setting axis labels
ax.set_ylabel('y values')
ax.set_xlabel('x values')
ax.set_title('My plot')
ax.grid(True)

# number of ticks
ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 1))

#plotting functions
y = 2*x**3 +1
plt.plot(x,y,'r',label='My cubic')
plt.plot([4],[6],'g',label='My point')
plt.plot(x,3*x,label='My line')
plt.legend()
plt.show()
```
