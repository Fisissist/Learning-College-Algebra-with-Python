# Scatter plot game

*   Randomly generate points on a graph and the player has to input the (x,y) coordinates
*   For added difficulty, make the graph larger


```python
import matplotlib.pyplot as plt
from random import randint
from numpy import arange


print("Welcome to the scatter plot game!")
print('Select a difficulty by entering a number betweern 1 and 3')
Difficulty = int(input("1. Easy\n2. Challenging\n3. Impossible\n"))

if Difficulty == 1:
    Size = 8
    ticks = 2
elif Difficulty == 2:
    Size = 16
    ticks = 4
elif Difficulty == 3:
    Size = 32
    ticks = 8
else:
    print("Invalid input")

xmin = -Size
xmax = Size
ymin = -Size
ymax = Size

fig, ax = plt.subplots()

x0 = randint(xmin, xmax)
y0 = randint(ymin, ymax)

plt.axis([xmin,xmax,ymin,ymax])
plt.plot([xmin,xmax],[0,0],'k')
plt.plot([0,0],[ymin,ymax], 'k')

ax.set_title('Scatter Plot Game')
ax.grid(True)

# number of ticks
ax.set_xticks(arange(xmin, xmax, ticks))
ax.set_yticks(arange(ymin, ymax, ticks))

plt.plot([x0], [y0], 'ro')

plt.show()

guess = input("Enter the coordinates of the red point point, seperate with a comma like so 4,6: \n")
guess_array = guess.split(",")
xguess = int(guess_array[0])
yguess = int(guess_array[1])

if xguess == x0 and yguess == y0:
    print("Correct!")
else:
    print("Incorrect!")
    print("The correct answer is: (" + str(x0) + "," + str(y0) + ")")
```

# Algebra practice game


*   Generate one-step and two-step problems with random integer values and the player has to input the answer
*   Use positive and negative values. For added difficulty, make the numbers larger



```python
from random import sample
from sympy import solve, symbols

print("Welcome to the Algebra Practice Game!")

#Solutions
def theSolutionator(equation):
    solution1 = solve(equation[0])
    solution2 = solve(equation[1])
    solution3 = solve(equation[2])
    solution4 = solve(equation[3])
    solution5 = solve(equation[4])
    solution6 = solve(equation[5])

    return solution1, solution2, solution3, solution4, solution5, solution6

def rounder(sol1, sol2, sol3, sol4, sol5, sol6):
    s1=round(sol1[0],2)
    s2=round(sol2[0],2)
    s3=round(sol3[0],2)
    s4=round(sol4[0],2)
    s5=round(sol5[0],2)
    s6=round(sol6[0],2)

    #print('')
    #print(s1)
    #print(s2)
    #print(s3)
    #print(s4)
    #print(s5)
    #print(s6)
    #print('')

    return s1, s2, s3, s4, s5, s6

def grader(equations):
    sol1, sol2, sol3, sol4, sol5, sol6 = theSolutionator(equations)

    s1, s2, s3, s4, s5, s6 = rounder(sol1, sol2, sol3, sol4, sol5, sol6)
    sol = [str(s1), str(s2), str(s3), str(s4), str(s5), str(s6)]

    print('')
    print('Enter your solutions. If a decimal, round to 2 decimal places and if')
    print('  there is no whole number then enter a zero before the decimal like so, 0.(value).')

    for i in range(6):
      ex = str(input(str(i+1)+'. x = '))
      if ex == sol[i]:
        print("Correct!")
      else:
        print("Incorrect!")
        print("The correct answer is: "+str(sol[i]))


# One Step Problems
def oneStep(coef):
    x = symbols('x')
    a = coef[0]
    b = coef[1]
    c = coef[2]

    y1 = a*x-b
    y2 = x+b-c
    y3 = x/b-c
    y4 = (a/b)*x - c
    y5 = a/b+c/b-x
    y6 = x+b-c/a

    print('')
    print('Solve for x:')
    print("a = "+str(a),"b = "+str(b),"c = "+str(c))
    print('1. ax = b')
    print('2. x+b = d')
    print('3. x/b = d')
    print('4. (a/b)x = c')
    print('5. a/b + d/b = x')
    print('6. x+b = d/a')

    equations = [y1, y2, y3, y4, y5, y6]
    grader(equations)

#Two Step Problems
def twoStep(coef):
    x = symbols('x')
    a = coef[0]
    b = coef[1]
    c = coef[2]

    y1 = a*x+b-c
    y2 = a*x+c-b*x
    y3 = a*(x+b)-c
    y4 = (a/b)*x+(a/b)*x-c
    y5 = x/a+b-c
    y6 = (x+a)/b-c

    print('')
    print('Solve for x:')
    print("a = "+str(a),"b = "+str(b),"c = "+str(c))
    print('1. ax+b=c')
    print('2. ax+c=bx')
    print('3. a(x+b)=c')
    print('4. (a/b)x+(a/b)x=c')
    print('5. x/a+b=c')
    print('6. (x+a)/b=c')

    equations = [y1, y2, y3, y4, y5, y6]
    grader(equations)

"""
Prompts
"""
# Difficulty
print('')
print('Select a difficulty by entering a number betweern 1 and 3')
Difficulty = int(input("1. Simple\n2. Challenging\n"))

if Difficulty == 1:
  coef = sample(range(-20, 20), 3)
elif Difficulty == 2:
  coef = sample(range(-80, 80), 3)
else:
  print("Invalid input")

# Problem Type
print('')
print('Select a problem type by entering a number betweern 1 and 3')
problem_type = int(input("1. One-step\n2. Two-step\n"))

if problem_type == 1:
  oneStep(coef)
elif problem_type == 2:
  twoStep(coef)
else:
  print("Invalid input")
```

# Projectile game


*   Display a "wall" with random height and location. Player has to move sliders to adjust a parabolic path to clear the wall
*   For added difficulty, make a second level where players enter a, b, and c without sliders



```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
from random import randint
from IPython.display import display

#Gives Vertex
def QuadraticVertex(a,b,c):
    vx = -b/(2*a)
    vy = a*vx**2 + b*vx +c
    return(vy)

#sets the wall distance and hieght
def wall():
    distance = randint(2,5)
    hieght = randint(2,10)

    print('')
    print("Distance to wall = ", distance)
    print("Height of wall = ", hieght)
    return(distance,hieght)

def Glider():
    print('Glider!')
    print('')
    print('Try to find at what initial velocity is needed tojust clear the wall.')
    dis,h = wall()
    walldistance = [dis, dis]
    wallhieght = [0,h]

    def glide(b):

        a = -4.9
        c = 0
        vy = QuadraticVertex(a,b,c)

        xmin = -2
        xmax = dis + 5
        ymin = -2
        ymax = vy + 5

        points = 4*(xmax-xmin)
        x = np.linspace(xmin,xmax,points)
        y = a*x**2 + b*x

        fig, ax = plt.subplots()
        plt.axis([xmin,xmax,ymin,ymax])
        plt.plot([xmin,xmax],[0,0],'k')
        plt.plot([0,0],[ymin,ymax],'k')

        ax.set_ylabel('Hieght')
        ax.set_xlabel('Distance')
        ax.set_title('Rocket!')
        ax.grid(True)

        plt.plot(x,y,'b',label='Rocket')
        plt.plot(walldistance,wallhieght,'brown',linewidth=10)
        plt.fill_between([xmin,xmax], [0,0], ymin, facecolor='brown')
        plt.show()

    interactive_plot = interactive(glide, b=(0,40))
    display(interactive_plot)

def Rockets():
    print('Rockets!')

    dis,h = wall()
    walldistance = [dis, dis]
    wallhieght = [0,h]

    a = float(input("Enter an acceleration: "))
    b = float(input("Velocity needed to clear the wall = "))
    c = float(input("What is your initial position? "))

    vy = QuadraticVertex(a,b,c)

    xmin = -2
    xmax = dis + 5
    ymin = -2
    ymax = vy + 5

    points = 4*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)
    y = a*x**2 + b*x+ c

    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax])
    plt.plot([xmin,xmax],[0,0],'k')
    plt.plot([0,0],[ymin,ymax],'k')

    ax.set_ylabel('Hieght')
    ax.set_xlabel('Distance')
    ax.set_title('Rocket!')
    ax.grid(True)

    ax.set_xticks(np.arange(xmin, xmax, 2))
    ax.set_yticks(np.arange(ymin, ymax, 2))

    plt.plot(x,y,'b',label='Rocket')
    plt.fill_between([xmin,0],[0,0],[c,c], facecolor='gray') # ledge for initial hieght
    plt.plot(walldistance,wallhieght,'brown',linewidth=10)
    plt.fill_between([xmin,xmax], [0,0], ymin, facecolor='brown')

    plt.show()

Inquery = input('Wanna play a game? (y/n):')
if Inquery == 'y':
    Difficulty = int(input("1. Easy\n2. Challenging\n"))
    if Difficulty == 1:
        Glider()
    elif Difficulty == 2:
        Rockets()
    else:
        print('Invalid input')
else:
    print('Goodbye')
```
